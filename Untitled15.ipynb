```python
# Data handling and numerical computations
import pandas as pd
import numpy as np

# Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

# Machine Learning - Model Training & Evaluation
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Handling Class Imbalance
from imblearn.over_sampling import SMOTE

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Load the dataset
df = pd.read_csv(r"D:\research paper\Personalized_Diet_Recommendations.csv")

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())

# Check dataset shape
print("\nDataset shape (rows, columns):", df.shape)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check data types
print("\nData types:")
print(df.dtypes.value_counts())
```

```python
# Add small random noise to numerical columns
df['Carbohydrate_Intake'] += np.random.normal(0, 20, df.shape[0])
df['Fat_Intake'] += np.random.normal(0, 15, df.shape[0])
```

```python
# Define diet categories
df['Smart_Diet'] = df.apply(
    lambda row: 'Low Carb' if row['Carbohydrate_Intake'] < 180 else
                'High Fat' if row['Fat_Intake'] > 90 else
                'Balanced', axis=1)

# Map to final diet plan names
df['Recommended_Meal_Plan'] = df['Smart_Diet'].map({
    'Low Carb': 'Keto',
    'High Fat': 'Paleo',
    'Balanced': 'Mediterranean'
})

# Drop intermediate column
df.drop(columns=['Smart_Diet'], inplace=True)
```

```python
df.drop(columns=['Patient_ID', 'BMI'], inplace=True)
```

```python
imputer = SimpleImputer(strategy='constant', fill_value='Unknown')
for col in ['Chronic_Disease', 'Allergies', 'Food_Aversions']:
    if col in df.columns and df[col].dtype == 'object':
        df[col] = imputer.fit_transform(df[[col]]).ravel()
```

```python
cat_cols = df.select_dtypes(include='object').columns.difference(['Recommended_Meal_Plan'])
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le  # Store encoders for future use
```

```python
le_rmp = LabelEncoder()
df['Recommended_Meal_Plan'] = le_rmp.fit_transform(df['Recommended_Meal_Plan'])
```

```python
X = df.drop(columns=['Recommended_Meal_Plan'])
y = df['Recommended_Meal_Plan']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
# imported warning function , because the joblib will removed from the new window version , so it can not find the phygical number
import warnings
warnings.filterwarnings("ignore", message="Could not find the number of physical cores")

#Actual project code--------------------------------------------------------------------------------------------------------------------------------------------


print("Original class distribution:", Counter(y_train))

# Apply SMOTE to balance classes
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

print("After SMOTE:", Counter(y_train_bal))
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_bal)
X_test_scaled = scaler.transform(X_test)
```

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X_train_scaled, y_train_bal)

# Visualize feature importance
importances = rf.feature_importances_
indices = np.argsort(importances)[-15:]  # Top 15 features
plt.figure(figsize=(12, 8))
plt.title('Top 15 Important Features', fontsize=14)
plt.barh(range(len(indices)), importances[indices], color='skyblue', align='center')
plt.yticks(range(len(indices)), [X.columns[i] for i in indices], fontsize=10)
plt.xlabel('Relative Importance', fontsize=12)
plt.show()

# Select top features
selector = SelectFromModel(rf, threshold="1.0*mean", prefit=True)
X_train_sel = selector.transform(X_train_scaled)
X_test_sel = selector.transform(X_test_scaled)
selected_features = X.columns[selector.get_support()].tolist()
print("Selected Features:", selected_features)
```

```python
import psutil
import os
import time
# Initialize process for memory tracking
process = psutil.Process(os.getpid())

def get_memory_usage():
    #Get current memory usage in MB
    return process.memory_info().rss / 1024 / 1024  # Convert to MB

models = {
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=4, random_state=42),
    "XGBoost": XGBClassifier(max_depth=3, learning_rate=0.07, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "SVM": SVC(kernel='rbf', C=1.0, random_state=42),
    "LightGBM": LGBMClassifier(num_leaves=20, learning_rate=0.05, random_state=42),
    "CatBoost": CatBoostClassifier(iterations=100, depth=3, learning_rate=0.05, verbose=0, random_state=42)
}
```

```python
# for slient the whole lightGBM warnings, i add warning codes
import warnings
warnings.filterwarnings("ignore")   # hide all warnings globally

from lightgbm import LGBMClassifier

# create LightGBM model with no console spam
model = LGBMClassifier(
    n_estimators=300,
    learning_rate=0.05,
    num_leaves=31,
    max_depth=-1,
    min_data_in_leaf=20,
    verbose=-1   # suppress LightGBM logs
)

# train the model
model.fit(X_train, y_train)

# predict (make sure X_test is a DataFrame if X_train was)
y_pred = model.predict(X_test)
  



# Actual code of project----------------------------------------------------------------------------------------------------------------------------

results = []
for name, model in models.items():
    print(f"Training {name}:-")
    
    # Track memory before training
    memory_before = get_memory_usage()
    
    # Track training time
    train_start_time = time.time()
    model.fit(X_train_sel, y_train_bal)
    training_time = time.time() - train_start_time
    
    # Track memory after training
    memory_after_training = get_memory_usage()
    training_memory = memory_after_training - memory_before
    
    # Track prediction time
    test_start_time = time.time()
    y_pred = model.predict(X_test_sel)
    prediction_time = time.time() - test_start_time
    
    # Track memory after prediction
    memory_after_prediction = get_memory_usage()
    total_memory = memory_after_prediction - memory_before
    
    # Calculate metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    cv = cross_val_score(model, X_train_sel, y_train_bal, cv=5, scoring='accuracy')
    
    results.append({
        'Model': name,
        'Accuracy': acc,
        'F1 Score': f1,
        'CV Accuracy Mean': cv.mean(),
        'CV Accuracy Std': cv.std(),
        'Training Time (s)': training_time,
        'Prediction Time (s)': prediction_time,
        'Total Execution Time (s)': training_time + prediction_time,
        'Training Memory (MB)': training_memory,
        'Total Memory (MB)': total_memory
    })
    
    # Confusion matrix for each model
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=le_rmp.classes_, 
                yticklabels=le_rmp.classes_)
    plt.title(f'Confusion Matrix - {name}', fontsize=12)
    plt.xlabel('Predicted', fontsize=10)
    plt.ylabel('Actual', fontsize=10)
    plt.show()

# Display results
results_df = pd.DataFrame(results)
results_df = results_df.sort_values('Accuracy', ascending=False)
print("\nModel Performance Comparison:")
print(results_df.to_string(index=False, float_format='%.4f'))

# Create visualizations for the new metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Training Time Comparison
results_df_sorted = results_df.sort_values('Training Time (s)', ascending=True)
axes[0, 0].barh(results_df_sorted['Model'], results_df_sorted['Training Time (s)'])
axes[0, 0].set_title('Training Time Comparison (seconds)')
axes[0, 0].set_xlabel('Time (seconds)')

# Prediction Time Comparison
results_df_sorted = results_df.sort_values('Prediction Time (s)', ascending=True)
axes[0, 1].barh(results_df_sorted['Model'], results_df_sorted['Prediction Time (s)'])
axes[0, 1].set_title('Prediction Time Comparison (seconds)')
axes[0, 1].set_xlabel('Time (seconds)')

# Memory Usage Comparison
results_df_sorted = results_df.sort_values('Total Memory (MB)', ascending=True)
axes[1, 0].barh(results_df_sorted['Model'], results_df_sorted['Total Memory (MB)'])
axes[1, 0].set_title('Total Memory Usage (MB)')
axes[1, 0].set_xlabel('Memory (MB)')

# Accuracy vs Time Comparison
axes[1, 1].scatter(results_df['Total Execution Time (s)'], results_df['Accuracy'])
for i, txt in enumerate(results_df['Model']):
    axes[1, 1].annotate(txt, (results_df['Total Execution Time (s)'].iloc[i], 
                             results_df['Accuracy'].iloc[i]))
axes[1, 1].set_title('Accuracy vs Total Execution Time')
axes[1, 1].set_xlabel('Total Execution Time (seconds)')
axes[1, 1].set_ylabel('Accuracy')

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nPerformance Metrics Summary:")
print(f"Fastest Training: {results_df.loc[results_df['Training Time (s)'].idxmin(), 'Model']} "
      f"({results_df['Training Time (s)'].min():.4f}s)")
print(f"Fastest Prediction: {results_df.loc[results_df['Prediction Time (s)'].idxmin(), 'Model']} "
      f"({results_df['Prediction Time (s)'].min():.4f}s)")
print(f"Most Memory Efficient: {results_df.loc[results_df['Total Memory (MB)'].idxmin(), 'Model']} "
      f"({results_df['Total Memory (MB)'].min():.2f}MB)")
print(f"Best Accuracy: {results_df.loc[results_df['Accuracy'].idxmax(), 'Model']} "
      f"({results_df['Accuracy'].max():.4f})")
```
